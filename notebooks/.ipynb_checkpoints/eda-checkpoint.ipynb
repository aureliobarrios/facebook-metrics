```python
#boxplots of our lifetime_impressions column by type and category column
fig, ax = plt.subplots(1, 2, figsize=(12.5, 10))
sns.boxplot(data=df, x='lifetime_impressions', y='type', ax=ax[0])
sns.boxplot(data=df, x='lifetime_impressions', y='category', ax=ax[1])
```

```python
#boxplots of our lifetime_consumers column by type and category column
fig, ax = plt.subplots(1, 2, figsize=(12.5, 10))
sns.boxplot(data=df, x='lifetime_consumers', y='type', ax=ax[0])
sns.boxplot(data=df, x='lifetime_consumers', y='category', ax=ax[1])
```

```python
#boxplots of our impressions_by_likers column by type and category column
fig, ax = plt.subplots(1, 2, figsize=(12.5, 10))
sns.boxplot(data=df, x='impressions_by_likers', y='type', ax=ax[0])
sns.boxplot(data=df, x='impressions_by_likers', y='category', ax=ax[1])
```

It is very evident from the plots above that our dataset is prone to outliers. There seems to be two events in our data where our impressions are above 600,000. Lets take a look at these rows two try and get a better understanding of these post events.

```python
df[df['impressions_by_likers'] > 600000]
```

## Getting A Better Understanding Of Our Outliers

In order to determine how to handle our outliers we must get a better understanding. From these two rows it seems to be that these two posts had a significant amount of impressions. Even though they had more impressions than other posts it seems that their other metrics like reach, engagement and consumptions are on par with other posts. 

```python
#save a dataframe of our outlier rows
outlier_df = df[df['impressions_by_likers'] > 600000]
```

```python
#get the lifetime_reach of both our impression outlier rows
outlier_reach = list(outlier_df['lifetime_reach'])

#plot the values of lifetime_reach for our outliers with the rest of the dataset
sns.displot(x=df['lifetime_reach'], kde=True)
plt.axvline(outlier_reach[0], color='red')
plt.axvline(outlier_reach[1], color='orange')
```

```python
#get the lifetime_engaged_users of both our impression outlier rows
outlier_engagement = list(outlier_df['lifetime_engaged_users'])

#plot the values of lifetime_engaged_users for our outliers with the rest of the dataset
sns.displot(x=df['lifetime_engaged_users'], kde=True)
plt.axvline(outlier_engagement[0], color='red')
plt.axvline(outlier_engagement[1], color='orange')
```

```python
#get the lifetime_consumers of both our impression outlier rows
outlier_consumers = list(outlier_df['lifetime_consumers'])

#plot the values of lifetime_consumers for our outliers with the rest of the dataset
sns.displot(x=df['lifetime_consumers'], kde=True)
plt.axvline(outlier_consumers[0], color='red')
plt.axvline(outlier_consumers[1], color='orange')
```

```python
#get the lifetime_consumptions of both our impression outlier rows
outlier_consumptions = list(outlier_df['lifetime_consumptions'])

#plot the values of lifetime_consumptions for our outliers with the rest of the dataset
sns.displot(x=df['lifetime_consumptions'], kde=True)
plt.axvline(outlier_consumptions[0], color='red')
plt.axvline(outlier_consumptions[1], color='orange')
```

As we can see from the plots above our two rows that have outlier values for the `impressions` columns, these same rows do not have outlier values in other columns like our consumptions, engagement and reach columns. It will be very important to see how large values of impressions can affect our other metrics. Therefore since these rows don't seem to be extremely out of the ordinary we should keep them in our analysis to get a better understanding of them. 

## Answering Questions w/ Initial Plots

Now that we have came to the conclusion that our outliers should stay we can move forward in our analysis. We begin our analysis by creating some initial plots that can help answer some initial questions we have about our data. 

The first question we can ask is how does our data look like by different post type. From our data we know that there are 4 different types of post types: `Photo`, `Status`, `Link` and `Video`.

```python
df['type'].value_counts()
```

```python
#distribution of likes by different post type
sns.histplot(x='total_likes', hue='type', data=df, multiple='dodge')
```

From our plots we notice that although our post are mainly `Photo` we can see that they all follow a similar distribution. 

Lets see how our posts differ when they are `Paid` or not paid. Do posts that are paid ads do worse than those posts that are directly from the page and what they want to display. Does the fact that a post is paid affect the engagement? Does this affect users that have already liked the page differently than users that have not liked the page?

```python
df['paid'].value_counts()
```

```python
plt.figure(figsize=(12.5, 10))
sns.histplot(x='lifetime_engaged_users', hue='paid', data=df, multiple='dodge')
```

```python
plt.figure(figsize=(12.5, 10))
sns.histplot(x='likers_who_engaged', hue='paid', data=df, multiple='dodge')
```

It seems like engagement is fairly similar in terms of the post being paid or not. This is also similar when we compare the distributions between users who have liked the page or not. It is important to understand that this is simply an eye test, from our graphs there doesn't seem to be an impact from wether a post was paid or not. In order to truly understand if a post being paid has an impact on other columns we would have to perform some tests which is something that can be done in the future but for the purpose of this project can be left as is. 

In the final step of our data analysis we should get familiar with our `total_interactions` column as this column will be the target variable of our machine learning model in the next step.  

```python
#create a boxplot of our total_interactions column
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='total_interactions')
```

```python
#create a correlation map of our features with our target variable
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(
    df.iloc[:, 7:].corr()[['total_interactions']].sort_values(by='total_interactions', ascending=False),
    annot=True
)

heatmap.set_title('Features Correlating with total_interactions')
```

We can use a correlation heatmap to understand how columns are correlated to our `total_interactions` target column. At the same time we can get a better understanding on how other columns are correlated to each other. In example we can see that the `lifetime_consumers` column is strongly correlated with our `lifetime_engaged_users` column. This observation is something that should be strongly considered when building our model. Another interesting observation is that all our columns are positively correlated with each other, there is no negative correlation present within our dataset. 

```python
#create a correlation heatmap of our numerical columns
plt.figure(figsize=(12.5, 10))
sns.heatmap(df.iloc[:, 7:].corr(), annot=True)
```

## Conclusions

From this analysis we understood that our dataset is prone to outliers. Altough our dataset has outliers we found that these outliers are necessary to truly understand our dataset in the next stages. This is because we found that the values that are outliers did not show to have a significant effect on other variables and ultimately our target variable which is the `total_interactions` column. This shows us that ultimately it is the mixture of other variables that ultimately decides how many interactions a post gets. 

We also got a glimpse at how a post being paid or not paid has an effect on our metrics. We found that our metrics follow a similar distribution being paid or not paid. We also found that most of our posts are photos, but the distributions for other types of posts follow the same distribution. 

Lastly we found how our attributes are correlated. We found that all our metrics are positively correlated which makes sense since our attributes complement each other. From a human perspective it makes sense that a post that has a lot of impressions also has a lot of engagement. It also makes sense that there are more engaged users when there are more consumers. This is very important to consider when building our model. Since most of our attributes are correlated with each other it could be useful to use some principal component analysis to better select our features and remove some of the redundant features in our dataset.
